```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
iris = pd.read_csv("/content/Iris.csv")
```

```python
iris
```

```python
feature_cols = iris[['Id', 'SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]

feature_cols
```

```python
label=iris[['Species']]
label
```

```python
iris[['Species']].isna().sum()
```

```python
iris['Species'].nunique()
```

```python
#split dataset in features and target variable

X = feature_cols # Features
y = label # Target variable
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.3, random_state=1)
```

```python
# Create Decision Tree classifer object
clf = DecisionTreeClassifier()
# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)
#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
# Let the user input new values for prediction
user_input = []
for feature in feature_cols:
    value = float(input(f"Enter the value for {feature}: "))
    user_input.append(value)

# Reshape the input to match the model's expectations
user_input = [user_input]

# Make a prediction on the user's input
prediction = clf.predict(user_input)
print(f"Predicted Outcome: {prediction[0]}")
```
